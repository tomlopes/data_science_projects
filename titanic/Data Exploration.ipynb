# Imports

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Import Data

```python
df = pd.read_csv('titanic.csv')
```

```python
df
```

```python
df.dtypes
```

```python
df['PassengerId'] = df['PassengerId'].astype('category')
df['Survived'] = df['Survived'].astype('category')
df['Pclass'] = df['Pclass'].astype('category')
df['Sex'] = df['Sex'].astype('category')
df['Embarked'] = df['Embarked'].astype('category')
```

```python
df.dtypes
```

# Dataset Summary

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.describe(include = 'all')
```

```python
df.isnull().sum()
```

# Dataset Distributions

```python
numerical_features = ['Age', 'SibSp', 'Parch', 'Fare']
categorical_features = ['Pclass', 'Sex', 'Cabin', 'Embarked']
target = 'Survived'
```

```python
sns.boxplot(df)
```

```python
for column in numerical_features:
    plt.figure()
    sns.boxplot(df[column])
    plt.title(f'Boxplot: {column}')
    plt.show()
```

```python
for column in numerical_features:
    sns.boxplot(data = df, x = column, y = target)
    plt.show()
```

```python
for column in numerical_features:
    df.hist(column)
```

```python
for column in numerical_features:
    df[column].plot(kind = 'kde')
    plt.title(f'Kernel Density Estimate: {column}')
    plt.xlabel(column)
    plt.tight_layout()
    plt.show()
```

# Correlations

```python
# correlation matrix for the numeric variables
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap='coolwarm', fmt='.2f')
```

```python
from scipy.stats import f_oneway

for column in numerical_features:
    groups = [df[df[target] == i][column] for i in df[target].unique()]
    f_stat, p_value = f_oneway(*groups)
    print(f"{column}: \nF-statistic: {f_stat:.10f}, P-value: {p_value:.10f}")
```

```python
numerical_columns = df.select_dtypes(include=['float64', 'int64'])

sns.pairplot(numerical_columns, kind = 'reg')
plt.show()
```

# Categorical Features

```python
for column in categorical_features:
    print(df[column].value_counts())
```

```python
for column in categorical_features:
    sns.countplot(df[column])
    plt.show()
```

```python
for column in categorical_features:
    sns.countplot(data=df, x=column, hue=target)
    plt.title(f'{column} Countplot with respect with the target')
    plt.show()
```
